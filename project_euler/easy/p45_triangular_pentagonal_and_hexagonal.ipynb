# Triangular, pentagonal, and hexagonal

Triangle, pentagonal, and hexagonal numbers are generated by the following formulae:

Triangle	 	Tn=n(n+1)/2	 	1, 3, 6, 10, 15, ...
Pentagonal	 	Pn=n(3n−1)/2	 	1, 5, 12, 22, 35, ...
Hexagonal	 	Hn=n(2n−1)	 	1, 6, 15, 28, 45, ...
It can be verified that T285 = P165 = H143 = 40755.

Write a python program to find the next triangle number that is also pentagonal and hexagonal.

## Results

```python
answer = 0 # Manually inserted for testing

def triangle(n):
    return n * (n+1) // 2

def pentagonal(n):
    return n * (3*n - 1) // 2

def hexagonal(n):
    return n * (2*n - 1)

# starting values for n
t = 286
p = 166
h = 144

# loop until a common value is found
while True:
    # calculate the triangle, pentagonal, and hexagonal numbers
    tn = triangle(t)
    pn = pentagonal(p)
    hn = hexagonal(h)

    # check if they are equal
    if tn == pn and pn == hn:
        print("The next triangle number that is also pentagonal and hexagonal is:", tn)
        answer = tn # manually inserted for testing
        break

    # increment the value of n for the series with the smallest value
    if tn <= pn and tn <= hn:
        t += 1
    elif pn <= tn and pn <= hn:
        p += 1
    else:
        h += 1
```
